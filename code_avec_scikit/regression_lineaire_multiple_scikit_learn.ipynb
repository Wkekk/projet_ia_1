```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
tp1 = pd.read_csv('boston_house_prices.csv')
tp1.head()
```

```python
from sklearn.preprocessing import StandardScaler
x = tp1.iloc[:, [5, 12]].values#.reshape(-1, 1)
#x = tp1.iloc[:, 5].values.reshape(len(tp1, 1)
#x1 = tp1.iloc[:, 12].values.reshape(-1, 1)
#x1 = tp1.iloc[:, 5].values.reshape(len(tp1, 1)                                  
y = tp1.iloc[:, -1].values.reshape(-1, 1)
print(x.shape)
ssc=  StandardScaler()
x=ssc.fit_transform(x)
y=ssc.fit_transform(y)
```

```python
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_test, y_test)
regressor.fit(X_train, y_train)
y_train
```

```python
y_pred_train = regressor.predict(X_train)
y_pred_test = regressor.predict(X_test)
y_pred_test
```

```python
print(X_train.shape, y_train.shape, y_pred_train.shape)
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
fig= plt.figure()
ax=fig.add_subplot(111,projection='3d')
plt.xlabel("X_train[:, [5, 12]]")
plt.ylabel("y_train")
ax.scatter(X_train[:,0],X_train[:,1],y_train)
ax.scatter(X_train[:,0],X_train[:,1],y_pred_train,color = 'yellow')
```

```python
mse = mean_squared_error(y_test, y_pred_test)
print(mse)
```
